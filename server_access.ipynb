# Access to records files and check audio


##  import libraries

```python
import paramiko
import os
import keyring
import sys
from contextlib import closing
import IPython.display as ipd  # To play sound in the notebook
```

## Connect to Server using SSH Key and password # 

```python
keyfile = os.path.expanduser('./private_sshkey/id_rsa')
#password = keyring.get_password('SSH', keyfile)
password = "raspberry"
key = paramiko.RSAKey.from_private_key_file(keyfile, password=password)

c = paramiko.SSHClient()
c.set_missing_host_key_policy(paramiko.AutoAddPolicy())
print ('connecting')
c.connect( hostname = "www.covid-record.com", username = "root", pkey = key )
print('connected')
```

## Download records file to local folder # 

```python
#os. getcwd() 
destdir='/Users/claire/Documents/YourScienceBC/Covid-19/Covid/records'
remotedir='covid-record/backend/uploads/20200520'
with closing(c.open_sftp()) as sftp:
    #print(os. getcwd())
    # cd into remote directory
    sftp.chdir(remotedir)
    # cd to local destination directory
    os.chdir(destdir)
    # download all files in it to destdir directory
    for filename in sftp.listdir():
        sftp.get(filename, filename)
```
